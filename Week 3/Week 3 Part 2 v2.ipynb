### Get data from Wiki Page

```python
raw = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
type(raw) # list
toronto = pd.DataFrame(raw[0])
toronto = toronto[toronto["Borough"] != "Not assigned"].reset_index(drop = True)
toronto = toronto.replace(r'/',', ', regex=True)
for index, row in toronto.iterrows():
    if row["Neighbourhood"] == "Not assigned":
        row["Neighbourhood"] = row["Borough"]
```

### Get coordinates

```python
df1 = toronto.loc[0:25]
df2 = toronto.loc[26:50]
df3 = toronto.loc[51:75]
df4 = toronto.loc[76:102]
```

```python
drop_unknown = []
geolocator = Nominatim(user_agent="ny_explorer")
for index, row in df1.iterrows():
    address = row['Neighbourhood'] + ', Toronto'
    try:
        location = geolocator.geocode(address)
        latitude = location.latitude
        longitude = location.longitude
        print('The geograpical coordinate of {} are {}, {}.'.format(address, latitude, longitude))
        df1.loc[index, 'Latitude'] = latitude
        df1.loc[index, 'Longitude'] = longitude
    except AttributeError:
        print('Address: {}, with index {} will be dropped'.format(address, index))
        drop_unknown.append(index)
```
